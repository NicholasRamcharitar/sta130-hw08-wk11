1. Accuracy
Definition: Measures the proportion of true results (both true positives and true negatives) in the population.

Example:
Application: Email spam filtering.
Why: Accuracy is useful when false positives (non-spam emails flagged as spam) and false negatives (spam emails not flagged) are equally important. For example, in a simple spam classifier, if 95% of emails are non-spam, a high accuracy rate might indicate good overall performance.
Limitation: When the dataset is imbalanced (e.g., very few spam emails compared to non-spam), accuracy can be misleading because the model might predict the majority class (non-spam) most of the time.


2. Sensitivity (Recall)
Definition: Measures the proportion of actual positives that are correctly identified.

Example:
Application: Disease screening (e.g., identifying cancer in medical imaging).
Why: High sensitivity is critical in medical testing because missing a true positive (e.g., a patient with cancer) could lead to dire consequences. Even if a few false positives occur, follow-up tests can clarify the diagnosis.
Limitation: High sensitivity might come at the cost of increased false positives.


3. Specificity
Definition: Measures the proportion of actual negatives that are correctly identified.

Example:
Application: Fraud detection in banking transactions.
Why: Specificity is important in identifying genuine transactions (negatives) because incorrectly flagging legitimate transactions as fraudulent (false positives) can erode customer trust and create unnecessary operational costs.
Limitation: High specificity might result in missing actual fraud cases (false negatives).


4. Precision
Definition: Measures the proportion of positive identifications that were actually correct.

Example:
Application: Targeted marketing campaigns (e.g., predicting likely customers for a product).
Why: High precision ensures that resources are focused on likely buyers, minimizing wasted effort and cost. For instance, if a campaign predicts 100 likely buyers, and only 50 of those predictions are correct, the precision would be 50%.
Limitation: Focusing solely on precision may reduce recall, meaning fewer true positives are identified.
